```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None


# Read in Data
df = pd.read_csv(r"C:\Users\cassi\Documents\Github Projects\Portfolio Python\movies.csv")
```

```python
# Check Data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Check Data types for columns

df.dtypes
```

```python
# Take year from release date column as YearCorrect

df['YearCorrect'] = df['released'].astype(str).str[:4]
```

```python
# Sort by highest grossing movies

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop duplicates in data

df.drop_duplicates()
```

```python
# Quick check for outliers in regards to Gross Earnings using a Boxplot

df.boxplot(column=['gross'])
```

```python
# View gross earnings according to movie rating

sns.stripplot(x="rating", y="gross", data=df)
```

```python
# Scatter plot budget vs gross with movie ratings to try view correlation

sns.relplot(x="budget", y="gross", hue="rating", data=df)
```

```python
# Scatter plot just budget vs gross to try view correlation

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Movie')

plt.show
```

```python
# Plot budget vs gross using Seaborn to try view correlation

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"green"})
```

```python
# Quick look at some numeric correlation

df.corr(method = 'pearson')
```

```python
# Show previous correlation as heatmap for easier visualisation

#Can change correlation to Kendall or Pearson by changing method = _____
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, cmap="Greens", annot=True)

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# To better work with data change columns to numeric

df_numerised = df

for col_name in df_numerised.columns:
    if(df_numerised[col_name].dtype == 'object'):
        df_numerised[col_name] = df_numerised[col_name].astype('category')
        df_numerised[col_name] = df_numerised[col_name].cat.codes
                                   
df_numerised
```

```python
# View correlation of numerised data as table

df_numerised.corr()
```

```python
# Show correlation of numerised data with heatmap

correlation_matrix = df_numerised.corr(method = 'pearson')

sns.heatmap(correlation_matrix, cmap="Greens", annot=True)

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Show breakdown of correlation links

correlation_mat = df_numerised.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sorted correlation into pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Show high correlation pairs

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
